1. Menentukan Data untuk Visualisasi

Data yang digunakan mencakup:

    CustomerID: Identitas pelanggan (akan diabaikan karena tidak relevan untuk visualisasi).
    Gender: Jenis kelamin pelanggan.
    Age: Usia pelanggan.
    Annual Income (k$): Pendapatan tahunan dalam ribuan dolar.
    Spending Score (1-100): Skor pengeluaran (1-100) yang mewakili kebiasaan belanja pelanggan.

2.Menentukan Tujuan, Pengguna, Fungsi, dan Tone

    Tujuan: Memvisualisasikan hubungan antara pendapatan tahunan, skor pengeluaran, usia, dan jenis kelamin pelanggan. Ini untuk membantu memahami pola pengeluaran berdasarkan faktor demografis.
    Pengguna: Manajer pemasaran atau pengambil keputusan bisnis yang tertarik dalam memahami profil pelanggan.
    Fungsi: Memberikan wawasan tentang segmen pelanggan berdasarkan pendapatan, usia, jenis kelamin, dan skor pengeluaran. Fungsi ini membantu dalam menentukan strategi pemasaran.
    Tone: Profesional, informatif, dan berbasis data, fokus pada pengambilan keputusan bisnis.

```
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```
df = pd.read_csv('Mall_Customers.csv')
df
```

```
# Menghitung jumlah setiap gender
gender_counts = df['Gender'].value_counts()

# Membuat pie chart untuk distribusi gender
plt.figure(figsize=(8, 6))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', colors=sns.color_palette('Set2'), startangle=90)
plt.title('Customer Count by Gender', fontsize=14)
plt.axis('equal')

# Menampilkan plot
plt.tight_layout()
plt.show()
```

```
# Set plot
sns.set(style="whitegrid")

plt.figure(figsize=(8, 6))
bar_plot = sns.barplot(
    x='Gender',
    y='Spending Score (1-100)',
    data=df,
    palette='Set2',
    ci=None
)

plt.title('Average Spending Score by Gender', fontsize=14)
plt.xlabel('Gender')
plt.ylabel('Spending Score (1-100)')

plt.tight_layout()
plt.show()
```

```
plt.figure(figsize=(8, 6))
box_plot = sns.boxplot(
    x='Gender',
    y='Annual Income (k$)',
    data=df,
    palette='Set2'
)

plt.title('Annual Income Distribution by Gender', fontsize=14)
plt.xlabel('Gender')
plt.ylabel('Annual Income (k$)')

plt.tight_layout()
plt.show()
```
